```python
import random
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
Choose="Ranger"

sns.set(style="whitegrid")

Class={"Barbarian":12,"Ranger":10,"Druid":8}
levels=7
ConMod=-1
InHp=Class[Choose]+ConMod
Hp=np.zeros(10000)+InHp
for i in range(2,levels+1):
    Increase = np.random.randint(1, Class[Choose],10000)
    print(min(Increase))
    mask = Increase == 1
    Reroll = np.random.randint(1, Class[Choose], 10000)[mask]
    Increase[mask] = Reroll
    Hp+=Increase+ConMod

Average=((Class[Choose]/2)+1+ConMod)*levels

plt.hist(Hp, bins=range(20, 90, 1), alpha=0.75)
text="Mean: "+str(np.mean(Hp))+" Median: "+str(np.median(Hp))+" Std: "+str(np.std(Hp))
print(text)
plt.title(str(Choose)+" HP at lvl "+str(levels)+" = "+str(np.mean(Hp)))
plt.axvline(np.mean(Hp), color='r', linestyle='dashed', linewidth=3,label="Rolled")
plt.axvline(Average, color='g', linestyle='dashed', linewidth=3,label="Ã„verage")
plt.axvline(np.median(Hp)+np.std(Hp), color='b', linestyle='dashed', linewidth=3,label="STD")
plt.axvline(np.median(Hp)-np.std(Hp), color='b', linestyle='dashed', linewidth=3)

plt.legend()
plt.xlabel("Average HP = "+str(Average))
plt.show()


```
