### Summon Define

Hit is a number added to a d20 

Damage will be a dictionary of number of 1d4, 1d6, 1d10, etc 

Damage Mod is a list that adds the mod to the roll

### Write Summon

Roll to Hit

Roll Damage

```python
Name="Velociraptor"
AttackNumber=2
Hit = [5,5]

D1  =    {"d4":0,"d6":2,"d8":0,"d10":0,"d12":0} 
D2  =    {"d4":3,"d6":0,"d8":0,"d10":0,"d12":0} 
D3  =    {"d4":0,"d6":0,"d8":0,"d10":0,"d12":0} 
D4  =    {"d4":0,"d6":0,"d8":0,"d10":0,"d12":0} 
Damage=[D1,D2,D3,D4]
DamageMod=[5,3]
Advantage=True
####################
#Write Summon
####################

S=Summon.Summon(Name,AttackNumber,Hit,Damage,DamageMod,Itnum=10000)

S.RollHit(hasAdvantage=Advantage,ShowPlot=True)
S.RollDamage(ShowPlot=True)
```

### Define Enemy

Load Roll to Hit

Load Roll Damage

Pass AC ?

Subtract HP

```python
Name="Zombie"
HP=50
AC=15
E=Enemy.Enemy(Name,HP,AC)
E.LoadHitRolls(S.HitRolls)
E.LoadDamageRolls(S.DamageRolls)
```
